```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
excel_file = 'data_ikan_3.xlsx'  # Ganti dengan nama file Excel Anda
sheet_name_1 = 'Aquarium_7'
sheet_name_2 = 'Aquarium_8'
sheet_name_3 = 'Aquarium_9'# Ganti dengan nama lembar kerja yang sesuai
df_1 = pd.read_excel(excel_file, sheet_name=sheet_name_1)
df_2 = pd.read_excel(excel_file, sheet_name=sheet_name_2)
df_3 = pd.read_excel(excel_file, sheet_name=sheet_name_3)
```

```python
data_aquarium_7 = df_1[df_1['Akuarium'] == 7]
data_aquarium_8 = df_2[df_2['Akuarium'] == 8]
data_aquarium_9 = df_3[df_3['Akuarium'] == 9]
```

```python
def regression_analysis(data):
    panjang = data['Panjang'].values.reshape(-1, 1)
    berat = data['Berat'].values

    model = LinearRegression()
    model.fit(panjang, berat)

    r_squared = r2_score(berat, model.predict(panjang))
    return model, r_squared
```

```python
def plot_regression(data, model, r_squared):
    plt.scatter(data['Panjang'], data['Berat'], color='blue', label='Data')
    plt.plot(data['Panjang'], model.predict(data['Panjang'].values.reshape(-1, 1)), color='red', label='Regression Line')
    plt.title(f'Regression Analysis (R-squared: {r_squared:.2f})')
    plt.xlabel('Panjang')
    plt.ylabel('Berat')
    plt.legend()
    plt.show()
```

```python
model_1, r_squared_1 = regression_analysis(data_aquarium_7)
model_2, r_squared_2 = regression_analysis(data_aquarium_8)
model_3, r_squared_3 = regression_analysis(data_aquarium_9)
```

```python
plot_regression(data_aquarium_7, model_1, r_squared_1)
plot_regression(data_aquarium_8, model_2, r_squared_2)
plot_regression(data_aquarium_9, model_3, r_squared_3)
```

```python
print("Koefisien determinasi (R-squared) untuk setiap akuarium:")
print("Akuarium 1:", r_squared_1)
print("Akuarium 2:", r_squared_2)
print("Akuarium 3:", r_squared_3)
```

```python
r_squared_avg = (r_squared_1 + r_squared_2 + r_squared_3) / 3
print("Rata-rata Koefisien determinasi (R-squared) dari ketiga akuarium:", r_squared_avg)
```

```python
max_r_squared = max(r_squared_1, r_squared_2, r_squared_3)
if max_r_squared == r_squared_1:
    print("Kesimpulan: Akuarium 1 memiliki perkembangan ikan yang paling signifikan.")
elif max_r_squared == r_squared_2:
    print("Kesimpulan: Akuarium 2 memiliki perkembangan ikan yang paling signifikan.")
else:
    print("Kesimpulan: Akuarium 3 memiliki perkembangan ikan yang paling signifikan.")
```

```python
threshold = 0.7
```

```python
count_significant = sum(r_squared > threshold for r_squared in [r_squared_1, r_squared_2, r_squared_3])
```

```python
percentage_significant = (count_significant / 3) * 100
print("Persentase akuarium dengan perkembangan signifikan:", percentage_significant, "%")
```
