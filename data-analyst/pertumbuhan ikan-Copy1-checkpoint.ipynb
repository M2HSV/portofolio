```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = {
    'akuarium_7': np.array([[1, 15.2], [2, 16.4], [4, 20.6]]),
    'akuarium_8': np.array([[1, 15.4], [2, 18.3], [4, 22.3]]),
    'akuarium_9': np.array([[1, 14.5], [2, 16.8], [4, 21.1]])
}
```

```python
koefisien_regresi = {}
```

```python
for akuarium, berat in data.items():
    X = berat[:, 0].reshape(-1, 1)  # Minggu ke
    y = berat[:, 1]  # Berat rata-rata

    model = LinearRegression()
    model.fit(X, y)

    koefisien_regresi[akuarium] = model.coef_[0]

    # Plot data
    plt.scatter(X, y, label=akuarium)
    
    # Plot garis regresi
    plt.plot(X, model.predict(X), color='red')

plt.xlabel('Minggu ke-')
plt.ylabel('Berat Rata-rata (gram)')
plt.title('Pertumbuhan Ikan dalam Akuarium')
plt.legend()
plt.grid(True)
plt.show()
```

```python
akuarium_tertinggi = max(koefisien_regresi, key=koefisien_regresi.get)
pertumbuhan_tertinggi = koefisien_regresi[akuarium_tertinggi]
```

```python
berat_awal = data[akuarium_tertinggi][0, 1]
persentase_pertumbuhan = (pertumbuhan_tertinggi / berat_awal) * 100
```

```python
print("Akuarium dengan pertumbuhan paling signifikan adalah:", akuarium_tertinggi)
print("Koefisien regresi (pertumbuhan) rata-rata:", pertumbuhan_tertinggi)
print("Persentase pertumbuhan:", round(persentase_pertumbuhan, 2), "%")
```
